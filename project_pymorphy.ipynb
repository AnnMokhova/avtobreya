```
import re
```

```
results = ''

with open('test.txt', 'r') as file:
  text = file.read()
  words = text.split()
  for word in words:
    word = re.sub(r'[^а-яА-я^-]', '', word)
    print(morph.parse(word)[0])
    results += str(morph.parse(word)[0])
```
